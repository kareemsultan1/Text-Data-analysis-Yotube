# 4..Collect Entire data of Youtube

```python
import os 
```

```python
path =r'D:\Yotube Analysis\1..Youtube\additional_data'
```

```python
files = os.listdir(path)
```

```python
files
```

```python
# we notice that the csv files come first and then json files aslo we need only a csv files
#so range = 1 as the first csv file , len(files) = y3dy 3l length l file kolo , 
#2 = The number 2 in this range means that the loop will skip every other item in the list. So it will print out the 
#indices 1, 3, 5, 7, etc.
# to split the csv file from the original file
for i in range(1,len(files),2):
    print(i)
```

```python
files_csv=[files[i] for i in range(0,len(files),2)]
```

```python
full_df = pd.DataFrame()
for file in files_csv:
    current_df=pd.read_csv(path + '/'+file,encoding='iso-8859-1',error_bad_lines=False)
    current_df['country']=file.split('.')[0][0:2]
    full_df=pd.concat([full_df,current_df])
```

# 6.. Find out whether audience is engaged or not

```python
full_df.columns
```

```python
full_df['likes_rate']=(full_df['likes'] / full_df['views'] * 100)
full_df['dislikes_rate']=(full_df['dislikes'] / full_df['views'] * 100)
full_df['comments_count_rate']=(full_df['comment_count'] / full_df['views'] * 100)
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='category_name',y='likes_rate',data=full_df)
plt.xticks(rotation='vertical')
plt.title("likes rate by categories")
plt.savefig("likes_rate by category name .jpeg")
```

# relation between views and likes

```python
plt.figure(figsize=(8,8))
sns.regplot(x='views',y='likes',data=full_df)
plt.xticks(rotation='vertical')
plt.title("relation between views and likes")
plt.savefig("relation between views and likes.jpeg")
```

# fining co-relation between (views,likes,dislikes)

```python
sns.heatmap(full_df[['views', 'likes', 'dislikes']].corr(),annot=True)
plt.savefig("co-relation between (views,likes,dislikes) .jpeg")
```
